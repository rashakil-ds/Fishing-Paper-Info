```python
x = df.drop('Phishing Status',axis=1)
y = df['Phishing Status']
```

# Over Sampling Technique

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
# https://imbalanced-learn.org/stable/references/generated/imblearn.over_sampling.RandomOverSampler.html
over = RandomOverSampler(random_state=100)
```

```python
x_new2 , y_new2 = over.fit_resample(x,y)
```

```python
#Seperate train and test data
from sklearn.model_selection import train_test_split as tts
xtrain2,xtest2, ytrain2,ytest2 = tts(x_new2, y_new2, train_size=.70,random_state=42) 
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
Clf = DecisionTreeClassifier()
```

```python
Clf.fit(xtrain2,ytrain2) 
```

```python
pred = Clf.predict(xtest2) #predicted class / y
pred
```

```python
import numpy as np
np.array(ytest2) # Actual class / y
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, plot_roc_curve
```

```python
print(classification_report(ytest2,pred))
```

```python
cm = confusion_matrix(ytest2, pred) #confusion matrix
```

```python
cm
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
```

```python
tp, fn, fp, tn = confusion_matrix(ytest2, pred).reshape(-1) #tp=true positive, fn=false negative, fp=false positive, tn=true negative
```

```python
score = (tp+tn) / (tp+ fn+ fp+ tn)
score
```

```python
Clf.score(xtest2,ytest2)
```

# ROC Curve & AUC

```python
plot_roc_curve(Clf,xtest2,ytest2,color='cyan')
plt.plot([0,1],[0,1])
plt.title('AUC Curve for Decision Tree')
```
